### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y_line = [i**2 for i in x]
bar_categories = ['A', 'B', 'C', 'D']
bar_values = [5, 7, 3, 9]
y_scatter = [2 * i for i in x]
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line chart
axs[0, 0].plot(x, y_line, marker='o')
axs[0, 0].set_title('Line Chart')

# Bar chart
axs[0, 1].bar(bar_categories, bar_values, color=['red', 'blue', 'green', 'orange'])
axs[0, 1].set_title('Bar Chart')

# Scatter plot
axs[1, 0].scatter(x, y_scatter, color='purple')
axs[1, 0].set_title('Scatter Plot')

# Pie chart
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart')
axs[1, 1].axis('equal')

# Adjust layout
plt.tight_layout()
plt.show()
```
